```python
import yfinance as yf
import numpy as np
import matplotlib.pyplot as plt
```

```python
''' PARAMS '''
ticker = 'AAPL'
start_date = '2021-03-03'
end_date = '2021-05-05'
```

```python
aapl = yf.Ticker(ticker)
```

```python
aapl_history = aapl.history(start=start_date, end=end_date)
```

```python
aapl_history
```

```python
data = aapl_history['Close'].to_numpy()
print(data.shape)
print(data)
```

```python
aapl_history['Close'].plot(figsize=(15,6), grid=True, legend=True)
aapl_history['Open'].plot(figsize=(15,6), grid=True, legend=True)
```
